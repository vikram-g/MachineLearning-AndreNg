```python
%matplotlib inline 
# Import modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('ex1data1.txt', header = None, names = ['population','profit'])
X = np.matrix(data.ix[:,0]).T
y = np.matrix(data.ix[:,1]).T
m = data.shape[0] # Number of rows
```

```python
#Scatter plot of population Vs Profit
plt.scatter(X,y, s = 40, c = 'r',marker = 'x')
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
```

```python
X = np.concatenate([np.matrix(np.ones(m)).T,X], axis = 1)
theta = np.matrix(np.zeros(2))
iterations = 1500
alpha = 0.01
```

```python
for i in range(iterations):
    theta = np.r_[theta,theta[i,:] - (alpha/m)*np.dot(X.T,(np.dot(X,theta[i,:].T) - y)).T]
    cost = sum(np.square(np.dot(X,theta[i,:].T) - y))/(2*m)
    print cost
```

```python
#Using normal equation
np.linalg.pinv((np.dot(X.T,X))).dot(X.T).dot(y).T
```

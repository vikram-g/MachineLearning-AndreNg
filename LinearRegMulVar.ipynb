```python
%matplotlib inline 
# Import modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import normalize
```

```python
data = pd.read_csv('ex1data2.txt', header = None, names = ['size','n_bedroom','price'])
X = np.matrix(data.ix[:,:-1])
y = np.matrix(data.ix[:,-1]).T

# Obtaining the mean and std to normalize any future values
mean_mat = np.mean(X,axis = 0)
std_mat = np.std(X,axis = 0)

#Normalize the X
X = normalize(X,axis = 0)

m = data.shape[0] # Number of rows
```

```python
X = np.concatenate([np.matrix(np.ones(m)).T,X], axis = 1)
theta = np.matrix(np.zeros(X.shape[1]))
iterations = 1500
alpha = 0.09
```

```python
for i in range(iterations):
    theta = np.r_[theta,theta[i,:] - (alpha/m)*np.dot(X.T,(np.dot(X,theta[i,:].T) - y)).T]
    cost = sum(np.square(np.dot(X,theta[i,:].T) - y))/(2*m)
    print cost
```
